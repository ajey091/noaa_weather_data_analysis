# NOAA Weather data

The NOAA (National Oceanic and Atmospheric Adminstration) aggregates world-wide, daily weather data and exposes it for researchers on an FTP server. They refer to the network of weather stations as the Global Historical Climatology Network. Here, we will download data for a few cities and perform some analysis. 

## Introduction

First we will download the data for a number of cities. 

We will choose 10 cities and get data for 10 years for the 10 cities.
The 10 chosen cities are the following:

             'Tokyo':'476620-99999',
             'Bangalore':'427056-99999',
             'Shanghai':'583620-99999',
             'Paris':'071570-99999',
             'Mexico_City':'766800-99999',
             'Cairo':'623660-99999',
             'Chicago':'725300-94846',
             'San_Francisco':'724940-23234',
             'Miami':'722040-12838',
             'Buenos_Aires':'875850-99999'

We get the data from 2013-2017 for these cities.

```python
import subprocess
import os
```

```python
locations = {'Tokyo':'476620-99999',
             'Bangalore':'427056-99999',
             'Shanghai':'583620-99999',
             'Paris':'071570-99999',
             'Mexico_City':'766800-99999',
             'Cairo':'623660-99999',
             'Chicago':'725300-94846',
             'San_Francisco':'724940-23234',
             'Miami':'722040-12838',
             'Buenos_Aires':'875850-99999'}


for year in range(2013,2018):
    subprocess.run(["mkdir", str(year)])
    os.environ['dir_name'] = str(year)
    os.chdir(os.environ['dir_name'])
    for city in locations:
        url = f'ftp://ftp.ncdc.noaa.gov/pub/data/noaa/isd-lite/{year}/{locations[city]}-{year}.gz'
        file_name = f'{city}_{year}.gz'
        print (url)
        subprocess.run(["curl", "-0", url, "--output", file_name])
        subprocess.run(["gunzip", file_name])
    os.chdir("..")
```

The above code creates folders for 5 years and downloads data for the 10 chosen cities. 


### Data Analysis

We will now visualize the locations of the 10 cities we have chosen for this analysis. We will use `geopandas` and `shapely` to plot the  locations on a map of the world. 

```python
import pandas as pd
import numpy as np
from shapely.geometry import Point
import geopandas as gpd
from geopandas import GeoDataFrame
import matplotlib.pyplot as plt


locations = {'Tokyo':'476620-99999',
             'Bangalore':'427056-99999',
             'Shanghai':'583620-99999',
             'Paris':'071570-99999',
             'Mexico_City':'766800-99999',
             'Cairo':'623660-99999',
             'Chicago':'725300-94846',
             'San_Francisco':'724940-23234',
             'Miami':'722040-12838',
             'Buenos_Aires':'875850-99999'}


isd_history = pd.read_csv('isd-history.csv')

lat = []
long = []

for city in locations:
    id = locations[city]
    id1,id2 = id.split('-')
    long.append(isd_history.loc[(isd_history['USAF'] == id1)]['LAT'].values[0])
    lat.append(isd_history.loc[(isd_history['USAF'] == id1)]['LON'].values[0])
    # print (lat,long)

df = pd.DataFrame(np.transpose(np.array([lat,long])),columns=['lat','long'])

geometry = [Point(xy) for xy in zip(df['lat'],df['long'])]
gdf = GeoDataFrame(df, geometry=geometry)

#this is a simple map that goes with geopandas
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
gdf.plot(ax=world.plot(figsize=(10, 6)), marker='o', color='red', markersize=20);
```

### Anomalous temperatures

We are defining anomalous temperatures as those temperatures that are >30 degrees Centigrade. Note that we chose the threshold of 28 degrees for Mexico city because there were years with no >30 degrees entries. We are interested in seeing the change in the frequency of anomalous temperatures over the time period. 

```python
anom_tokyo = tokyo_data[tokyo_data['temp'] > 300].groupby('year')['temp'].count()
anom_bangalore = bangalore_data[bangalore_data['temp'] > 300].groupby('year')['temp'].count()
anom_chicago = chicago_data[chicago_data['temp'] > 300].groupby('year')['temp'].count()
anom_miami = miami_data[miami_data['temp'] > 300].groupby('year')['temp'].count()
anom_buenos = buenos_data[buenos_data['temp'] > 300].groupby('year')['temp'].count()


anom_mexico = mexico_data[mexico_data['temp'] > 280].groupby('year')['temp'].count()
anom_cairo = cairo_data[cairo_data['temp'] > 300].groupby('year')['temp'].count()
anom_sf = sf_data[sf_data['temp'] > 300].groupby('year')['temp'].count()
anom_shanghai = shanghai_data[shanghai_data['temp'] > 300].groupby('year')['temp'].count()
anom_paris = paris_data[paris_data['temp'] > 300].groupby('year')['temp'].count()


anom_df_1 = pd.DataFrame({'year':anom_tokyo.index, 
                        'tokyo_freq':anom_tokyo.values,
                        'bangalore_freq':anom_bangalore.values,
                        'chicago_freq':anom_chicago.values,
                        'miami_freq':anom_miami.values,
                        'buenos_freq':anom_buenos.values,
                       })

anom_df_2 = pd.DataFrame({'year':anom_mexico.index, 
                        'mexico_freq':anom_mexico.values,
                        'cairo_freq':anom_cairo.values,
                        'sf_freq':anom_sf.values,
                        'shanghai_freq':anom_shanghai.values,
                        'paris_freq':anom_paris.values,
                       })
```

```python
fig, ax = plt.subplots(figsize=(16, 8))
anom_df_1.plot(x='year',kind='bar',ax=ax)
plt.title('Frequency of anomalous temperatures over time')
plt.ylabel('Number of anomalous temperature entries')
plt.grid()
ax.legend(["Tokyo", "Bangalore", "Chicago", "Miami", "Buenos Aires"]);
```

```python
fig, ax = plt.subplots(figsize=(16, 8))
anom_df_2.plot(x='year',kind='bar',ax=ax)
plt.title('Frequency of anomalous temperatures over time')
plt.ylabel('Number of anomalous temperature entries')
plt.grid()
ax.legend(["Mexico City", "Cairo", "San Francisco", "Shanghai", "Paris"]);
```

It certainly does seem like tropical cities are observing larger increases in anomalous temperatures. We will finally look at the maximum and minimum temperatures for these cities.
